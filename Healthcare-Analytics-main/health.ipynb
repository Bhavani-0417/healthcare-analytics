```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Step 1: Generate Synthetic Data
X, y = make_classification(
    n_samples=32000, 
    n_features=17, 
    n_classes=11, 
    n_informative=17,
    n_redundant=0,
    n_repeated=0,# Increase informative features
    n_clusters_per_class=2,  # Keep clusters per class default (or adjust if needed)
    random_state=42
)

# Convert to DataFrame
df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(X.shape[1])])
df['Target'] = y  # Add target column

# Step 2: Split into train and test sets
train, test1 = train_test_split(df, test_size=0.2, random_state=42)

# Step 3: Separate features and target
X_train = train.iloc[:, :-1]  # Features
y_train = train.iloc[:, -1].values  # Target as NumPy array

X_test = test1.iloc[:, :-1]    
y_test = test1.iloc[:, -1].values  # Target as NumPy array

# Step 4: Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 5: Train the XGBoost model
model = xgb.XGBClassifier()
model.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Step 2: Preprocessing (Assuming the last column is the target)
X_train = train.iloc[:, :-1]  # Features
y_train = train.iloc[:, -1]   # Target

X_test = test1.iloc[:, :-1]    
y_test = test1.iloc[:, -1]

# Step 3: Standardize the data (if necessary)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 4: Train the model (Using Random Forest)
model = RandomForestClassifier(
    n_estimators=200,  # Number of trees
    max_depth=5,       # Maximum depth of each tree
    random_state=42,   # Ensuring reproducibility
    n_jobs=-1          # Use all available CPU cores
)
model.fit(X_train, y_train)

# Step 5: Make predictions on test data
y_pred = model.predict(X_test)

# If the test.csv has actual labels, evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

# Use MinMaxScaler instead of StandardScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train the Random Forest with better hyperparameters
model1 = RandomForestClassifier(
    n_estimators=300,  # More trees for better performance
    max_depth=10,  # Increase depth for capturing more patterns
    min_samples_split=5,  # Prevent overfitting
    min_samples_leaf=2,  # Minimum samples per leaf
    max_features='sqrt',  # Consider sqrt of features per split
    random_state=42,
    n_jobs=-1
)
model1.fit(X_train, y_train)

# Predictions
y_pred = model1.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print(f"Improved Model Accuracy: {accuracy:.2f}")
```
